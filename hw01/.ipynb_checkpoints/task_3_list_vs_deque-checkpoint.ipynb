### Домашнее задание №3 [0 - 0.1 балла]

**Условие:**
    
Сравнить скорость работы и потребление памяти list и deque.

**Комментарий:**

Количесвто баллов зависит от полноты сравнения.

```python
import sys

from datetime import datetime
from collections import deque

import matplotlib.pyplot as plt
from memory_profiler import memory_usage
```

```python
# общие методы списка и deque

list_methods = dir(list())
deque_methods = dir(deque())

common_methods = set(deque_methods)&set(list_methods)
common_methods = sorted( m for m in common_methods if not m.startswith("_"))
print(common_methods)
```

### 1. Append

```python
l = list()
q = deque()

n = 10000

runtime_array = list()
runtime_deque = list()

for i in range(n):
    start = datetime.now().timestamp()
    l.append(i)
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(n):
    start = datetime.now().timestamp()
    q.append(i)
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0,0.0001)
plt.legend();
```

В среднем время работы append у листа больше чем у очереди, хотя и у тех и у тех, бывают выборосы вверх.

```python
n = 5000

l = list()
q = deque()

array_memory = list()
deque_memory = list()

for i in range(n):
    start = sys.getsizeof(l)
    l.append(i)
    end = sys.getsizeof(l)
    array_memory.append(end-start)
    
for i in range(n):
    start = sys.getsizeof(q)
    q.append(i)
    end = sys.getsizeof(q)
    deque_memory.append(end-start)
```

```python
# сравнение прироста занимаемой памяти 
plt.plot(array_memory, label="array")
plt.plot(deque_memory, label="deque")
plt.legend();
```

Список выделяет память с линейной зависимостью от своего объема, в то время как deque выделяет всегда одно и то же количество памяти.

### 2. Clear

```python
n = 10000

runtime_array = list()
runtime_deque = list()

for i in range(n):
    l = list(range(i))
    start = datetime.now().timestamp()
    l.clear()
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(n):
    q = deque(range(i))
    start = datetime.now().timestamp()
    q.clear()
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
    
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0,0.0001)
plt.legend();
```

Clear выполняется за линейное время

### 3. Copy

```python
n = 10000

runtime_array = list()
runtime_deque = list()

for i in range(n):
    l = list(range(i))
    start = datetime.now().timestamp()
    l_tmp = l.copy()
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(n):
    q = deque(range(i))
    start = datetime.now().timestamp()
    q_tmp = q.copy()
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
    
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0, 0.0005)
plt.legend();
```

Copy тоже выполняет за линейное время

### 4. Count

```python
n = 10000

runtime_array = list()
runtime_deque = list()

for i in range(n):
    l = list(range(i))
    start = datetime.now().timestamp()
    l.count(0)
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(n):
    q = deque(range(i))
    start = datetime.now().timestamp()
    q.count(0)
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
    
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0, 0.0005)
plt.legend();
```

Count обязательно дложен пробежать по всем элементам, поэтому тоже выполняется за линейное время.

### 5. Extend

```python
# к пустому списку будет добавлять массивы разной длины

l = list()
q = deque()

n = 10000

runtime_array = list()
runtime_deque = list()

for i in range(n):
    l_i = l.copy()
    new_list = list(range(i))
    start = datetime.now().timestamp()
    l_i.extend(new_list)
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(n):
    q_i = q.copy()
    new_deque = deque(range(i))
    start = datetime.now().timestamp()
    q_i.extend(new_deque)
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0, 0.0002)
plt.legend();
```

Обе структуры добавляют новые элементы за линейное время, но у deque оно растет быстрее.

```python
# к увеличивающимуся списку будет добавлять массивы одинаковой длины

l = list()
q = deque()

n = 10000

runtime_array = list()
runtime_deque = list()

new_list = list(range(2000))
new_deque = deque(range(2000))

for i in range(n):
    l.append(i)
    l_i = l.copy()
    start = datetime.now().timestamp()
    l_i.extend(new_list)
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(n):
    q.append(i)
    q_i = q.copy()
    start = datetime.now().timestamp()
    q_i.extend(new_deque)
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0, 0.0002)
plt.legend();
```

Время выполнения заметно сокращается у списка после того, как его длина становится больше длины добавляемого списка, в то время как у очереди это время практически не меняется.

### 6. index

```python
# лучший случай - элемент в начале списка

n = 10000

runtime_array = list()
runtime_deque = list()

for i in range(1,n):
    l = list(range(i))
    start = datetime.now().timestamp()
    l.index(0)
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(1,n):
    q = deque(range(i))
    start = datetime.now().timestamp()
    q.index(0)
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0, 0.00002)
plt.legend();
```

```python
# худший случай - элемент в конце списка

n = 10000

runtime_array = list()
runtime_deque = list()

for i in range(1,n):
    l = list(range(i))
    start = datetime.now().timestamp()
    l.index(i-1)
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(1,n):
    q = deque(range(i))
    start = datetime.now().timestamp()
    q.index(i-1)
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0, 0.0002)
plt.legend();
```

Как и ожидалось в лучшем случае поиск выполняется за O(1), а в худшем O(n).

### 7. Insert

```python
# лучший случай - элемент вставляется в конец списка

n = 10000

runtime_array = list()
runtime_deque = list()

for i in range(1,n):
    l = list(range(i))
    start = datetime.now().timestamp()
    l.insert(i-1, 0)
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(1,n):
    q = deque(range(i))
    start = datetime.now().timestamp()
    q.insert(i-1, 0)
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0, 0.00002)
plt.legend();
```

Время выполнения - константное, хотя у списка значительно больше вначале.

```python
# худший случай - элемент вставляется в начало списка

n = 100000

runtime_array = list()
runtime_deque = list()

for i in range(1, n, 100):
    l = list(range(i))
    start = datetime.now().timestamp()
    l.insert(0, 0)
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(1, n, 100):
    q = deque(range(i))
    start = datetime.now().timestamp()
    q.insert(0, 0)
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0, 0.0002)
plt.legend();
```

list - линейное время, deque - константное, т.к. list перезаписывает все ссылки

### 8. Pop

```python
# лучший случай - элемент удаляется из конца списка

n = 100000

runtime_array = list()
runtime_deque = list()

for i in range(1, n, 100):
    l = list(range(i))
    start = datetime.now().timestamp()
    l.pop()
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(1, n, 100):
    q = deque(range(i))
    start = datetime.now().timestamp()
    q.pop()
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0, 0.0001)
plt.legend();
```

```python
# худщий случай - элемент удаляется из начала списка

n = 100000

runtime_array = list()
runtime_deque = list()

for i in range(1, n, 100):
    l = list(range(i))
    start = datetime.now().timestamp()
    l.pop(0)
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(1, n, 100):
    q = deque(range(i))
    start = datetime.now().timestamp()
    q.popleft()
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0, 0.0003)
plt.legend();
```

list - линейное, т.к. перезаписывает элементы, deque - константное, т.к. просто перезаписывает ссылки.

### 9. Remove

```python
# удаляем первый элемент

n = 100000

runtime_array = list()
runtime_deque = list()

for i in range(1, n, 100):
    l = list(range(i))
    start = datetime.now().timestamp()
    l.remove(0)
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(1, n, 100):
    q = deque(range(i))
    start = datetime.now().timestamp()
    q.remove(0)
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0, 0.0003)
plt.legend();
```

```python
# удаляем последний элемент

n = 100000

runtime_array = list()
runtime_deque = list()

for i in range(1, n, 100):
    l = list(range(i))
    start = datetime.now().timestamp()
    l.remove(i-1)
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(1, n, 100):
    q = deque(range(i))
    start = datetime.now().timestamp()
    q.remove(i-1)
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0, 0.006)
plt.legend();
```

list - в обоих случаях дает линейное время, т.к. в первом случае, нужн оперезаписать все индексы, а во втором перебрать все элементы, до того, как он найдет нужный. deque - если элемент находится в начале, то операция выполняется за константное время, иначе за линейное.

### 10. Reverse

```python
n = 100000

runtime_array = list()
runtime_deque = list()

for i in range(1, n, 100):
    l = list(range(i))
    start = datetime.now().timestamp()
    l.reverse()
    end = datetime.now().timestamp()
    runtime_array.append(end-start)
    
for i in range(1, n, 100):
    q = deque(range(i))
    start = datetime.now().timestamp()
    q.reverse()
    end = datetime.now().timestamp()
    runtime_deque.append(end-start)
```

```python
plt.plot(runtime_array, label="array")
plt.plot(runtime_deque, label="deque")
plt.ylim(0, 0.0006)
plt.legend();
```

Обе структуры выполняют процедуру за линейное время.
